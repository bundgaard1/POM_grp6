```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = 0
env_q= 1
env_q_vel = 2
#time * torque_motor * tourque_grav * torque_fric 
motor_time = 0
torque_motor = 1
torque_grav = 2
torque_fric = 3
```

```python
Scenario = 3
datafile = f"{Scenario}.txt"
motordata = f"{Scenario}motor.txt"

input = np.loadtxt(datafile,dtype='f', delimiter=' ')
motorinput = np.loadtxt(motordata,dtype='f', delimiter=' ')
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(input[:,time], input[:,env_q], label="Robot arm angle")
plt.plot(input[:,time], input[:,env_q_vel], label="Robot arm angle velocity")
plt.plot(motorinput[:,motor_time], motorinput[:,torque_motor], label="Motor torque")

plt.legend()

plt.subplot(1,2,2)
plt.plot(motorinput[:,motor_time], motorinput[:,torque_fric], label="Friction torque")
plt.plot(motorinput[:,motor_time], motorinput[:,torque_grav], label="Gravity torque")

plt.legend()
```
